```python
counties = df.RegionName.unique()
if 'Tuolumne County' in counties:
    counties = np.delete(counties, np.where(counties == 'Tuolumne County'))
lag = 0
corr_case_rate = pd.DataFrame({'lag':lag, 'county': counties, 'corr':np.nan, 'pvalue':np.nan})

for county in counties:
    #df_county = df[df['RegionName'] == county].copy()
    if county != 'Tuolumne County':
        df_county = df[df['RegionName'] == county].copy()
    date = list(df_county['Date'].values)
    price = list(df_county['Price'].values)
    cases_rate = list(df_county['cases_rate'].values)
    death_rate = list(df_county['death_rate'].values)
    price_change = [0] * len(price)
    for i in range(1, len(price)):
        price_change[i] = price[i] - price[i-1]
    r = stats.pearsonr(price[25 + lag:], cases_rate[25:50-lag])
    #print(county.ljust(27,' '), '%.4f'%(r[0]))
    corr_case_rate.loc[corr_case_rate['county'] == county, 'corr'] = r[0]
    corr_case_rate.loc[corr_case_rate['county'] == county, 'pvalue'] = r[1]
    
corr_case_rate = corr_case_rate.sort_values(by = 'pvalue')
corr_case_rate.reset_index(inplace = True, drop = True)
corr_case_rate
```

```python
counties = df.RegionName.unique()
if 'Tuolumne County' in counties:
    counties = np.delete(counties, np.where(counties == 'Tuolumne County'))
lag = 0
corr_case_rate = pd.DataFrame({'lag':lag, 'x1':'price', 'x2':'case_rate', \
                               'county': counties, 'corr':np.nan, 'pvalue':np.nan})

for county in counties:
    #df_county = df[df['RegionName'] == county].copy()
    if county != 'Tuolumne County':
        df_county = df[df['RegionName'] == county].copy()
    date = list(df_county['Date'].values)
    price = list(df_county['Price'].values)
    cases_rate = list(df_county['cases_rate'].values)
    death_rate = list(df_county['death_rate'].values)
    price_change = [0] * len(price)
    for i in range(1, len(price)):
        price_change[i] = price[i] - price[i-1]
    r = stats.pearsonr(price[25 + lag:], cases_rate[25:50-lag])
    #print(county.ljust(27,' '), '%.4f'%(r[0]))
    corr_case_rate.loc[corr_case_rate['county'] == county, 'corr'] = r[0]
    corr_case_rate.loc[corr_case_rate['county'] == county, 'pvalue'] = r[1]
    
corr_case_rate = corr_case_rate.sort_values(by = 'pvalue')
corr_case_rate.reset_index(inplace = True, drop = True)
corr_case_rate
corr_describe0 = pd.DataFrame(columns = ['lag','describe','p-value<0.05', 'p-value<0.1'])
corr_describe0.loc[0, 'lag'] = 0
corr_describe0.loc[0, 'describe'] = 'correlation of housing price and case rate'
corr_describe0.loc[0, 'p-value<0.05'] = corr_case_rate[corr_case_rate['pvalue']<0.05].shape[0]
corr_describe0.loc[0, 'p-value<0.1'] = corr_case_rate[corr_case_rate['pvalue']<0.10].shape[0]
corr_describe0
```

## lagged correlation analysis

```python
def lag_corr(df, county):
    df_county = df[df['RegionName'] == county].copy()
    date = list(df_county['Date'].values)
    price = list(df_county['Price'].values)
    cases_rate = list(df_county['cases_rate'].values)
    death_rate = list(df_county['death_rate'].values)
    price_change = [0] * len(price)
    for i in range(1, len(price)):
        price_change[i] = price[i] - price[i-1]
    print('County:',county)
    

    #price and case rate
    plt.figure(figsize = (10, 5))
    plt.plot(date, price, 'b-', label = 'Price')
    plt.twinx()
    plt.plot(date, cases_rate, c = 'red', linestyle = '-', label = 'Death Rate')
    plt.xticks(rotation = 90)
    plt.legend()
    plt.show()
    
    #price change and case rate
    plt.figure(figsize = (10, 5))
    plt.plot(date, price_change, 'b-', label = 'Price Change')
    plt.twinx()
    plt.plot(date, cases_rate, c = 'red', linestyle = '-', label = 'Cases Rate')
    plt.xticks(rotation = 90)
    plt.legend()
    plt.show()
    
    #price and death rate
    plt.figure(figsize = (10, 5))
    plt.plot(date, price, 'b-', label = 'Price')
    plt.twinx()
    plt.plot(date, death_rate, c = 'red', linestyle = '-', label = 'Death Rate')
    plt.xticks(rotation = 90)
    plt.legend()
    plt.show()
    
    #price change and case rate
    plt.figure(figsize = (10, 5))
    plt.plot(date, price_change, 'b-', label = 'Price Change')
    plt.twinx()
    plt.plot(date, death_rate, c = 'red', linestyle = '-', label = 'Cases Rate')
    plt.xticks(rotation = 90)
    plt.legend()
    plt.show()
    
    print('======cases lag and price===')
    for lag in range(10):
        r = stats.pearsonr(price[25 + lag:], cases_rate[25:50-lag])
        print('lag of {} month, r = '.format(lag), r)
    
    print('\n\n======cases lag and price change===')

    for lag in range(10):
        r = stats.pearsonr(price_change[25 + lag:], cases_rate[25:50-lag])
        print('lag of {} month, r = '.format(lag), r)
    
    print('\n\n======death lag and price===')
    for lag in range(10):
        r = stats.pearsonr(price[25 + lag:], death_rate[25:50-lag])
        print('lag of {} month, r = '.format(lag), r)
    
    print('\n\n======death lag and price change===')

    for lag in range(10):
        r = stats.pearsonr(price_change[25 + lag:], death_rate[25:50-lag])
        print('lag of {} month, r = '.format(lag), r)
```

## lag of 6 months

```python
counties = df.RegionName.unique()
lag_corr_all = dict()
for county in counties:
    if county != 'Tuolumne County':
        df_county = df[df['RegionName'] == county].copy()
        
        lag = 6
        price = list(df_county['Price'].values)
        death_rate = list(df_county['death_rate'].values)

        price_change = [0] * len(price)
        for i in range(1, len(price)):
            price_change[i] = price[i] - price[i-1]
            
        r = stats.pearsonr(price_change[25 + lag:], death_rate[25:50-lag])
        lag_corr_all[county] = r
lag_corr_all
```
